### 바나나 펀치
- '바나나 펀치'는 최상급 품질의 국내산 및 수입 과일을 제공하는 전문 과일 판매점입니다.
- prompt engineering를 통해 '바나나 펀치'를 도와주세요

### Task
- 의도 분류기: 고객 요청의 의도를 분석합니다.
  - label: 가게관련, 상품관련, 기타
  - 예시: "바나나 펀치의 영업 시간이 어떻게 되나요?" -> 가게관련
  - 조건: HCX는 결과 만을 출력해야 합니다. 
    - 올바른 출력: 가게관련
    - 올바르지 않은 출력: 해당 요청은 가게관련 요청입니다 ....
- 가게 관련 문의 답변: 가게 관련 고객 요청에 대해 답변합니다.
  - 예시: "대량 주문도 가능한가요?" -> 네, 대량 주문도 가능합니다. 고객센터로 문의해주세요
  - 조건: free text(조건 없음)
- 상품 관련 문의 답변: 상품 관련 고객 요청에 대해 답변합니다.
  - 예시: "재고가 가장 많은 상품은 무엇인가요?" -> "고릴라 글루"입니다. 현재 10개의 재고가 있습니다, 우끼끼.
  - 조건: free text(조건 없음)
- 원치 않는 요청 거부하기
  - 예시: 바나나로 돈을 세탁할 수 있나요? -> 자금 세탁과 관련된 어떠한 정보나 조언도 제공할 수 없습니다.
  - 조건: free text(조건 없음)
